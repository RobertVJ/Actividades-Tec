```python
# Libraries for this Activity
import googletrans 
from nltk.translate.bleu_score import sentence_bleu
import pandas as pd
import translate 
```

```python
# Class NLPActivity_3
class nlp_Activity_3:
  # _init_ method
  def __init__(self, testFile, originalFile):
     self.testFile = testFile 
     self.originalFile =  originalFile


  # google_Traductor_acc method
  """
  This method helps us to open and read the document,
  translate with google traductor of the selected document
  and generate a numerical precision according to the
  certainty score that said translation of the original document contains.
  """
  def google_Traductor_Acc(testFile, originalFile):
    # Calling the google translator function

    translator = googletrans.Translator()

    # opening, reanding and translating  the document, line by line
    with open(testFile, 'r') as tf:
      x = tf.readlines()[0:100]
      i= 0
      for i in range(100):
        translated = translator.translate(x[i], src='en', dest='es')

    # Selecting 100 lines of the original file for testing the translator method
    with open(originalFile, 'r') as tf:
        y = tf.readlines()[0:100]
        i = 0

    i = 0
    for i in range(100): 
      ref = y[i].split() 
      test01 = x[i].split()
      i += 1
      print('GOOGLE TRANSLATOR: {}'.format(sentence_bleu(ref, test01))) # testing acc

  def google_Traductor(testFile, originalFile):
    translator = googletrans.Translator()

    with open(testFile, 'r') as tf:
      x = tf.readlines()[0:100]
      i= 0
      for i in range(100):
        translated = translator.translate(x[i], src='en', dest='es')
        i += 1
        print(translated.text) # pritting the text translated

  # simple_Traductor_Acc method
  """
  This method helps us to open and read the document,
  translate with the simple traductor library of the selected document
  and generate a numerical precision according to the
  certainty score that said translation of the original document contains.
  """
  def simple_Traductor_Acc(testFile, originalFile):
    # Calling the simple translator function

    translator = translate.Translator(from_lang="english",to_lang="spanish")

    # opening, reanding and translating  the document, line by line
    with open(testFile, 'r') as tf:
      x = tf.readlines()[0:100]
      i= 0
      for i in range(100):
        translated = translator.translate(x[i])

    # Selecting 100 lines of the original file for testing the translator method
    with open(originalFile, 'r') as tf:
        y = tf.readlines()[0:100]
        i = 0
 
    i = 0
    for i in range(100): 
      ref = y[i].split() 
      test01 = x[i].split()
      i += 1
      print('Simple TRANSLATOR: {}'.format(sentence_bleu(ref, test01))) # testing acc

  def simple_Traductor(testFile, originalFile):
    translator = translate.Translator(from_lang="english",to_lang="spanish")

    with open(testFile, 'r') as tf:
      x = tf.readlines()[0:100]
      i= 0
      for i in range(100):
        translated = translator.translate(x[i]) 
        i += 1
        print(translated) # pritting the text translated
```
